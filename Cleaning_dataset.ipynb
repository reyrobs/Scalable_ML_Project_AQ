###The Air_quality dataset:

```python
url = 'https://github.com/reyrobs/Scalable_ML_Project_AQ/raw/main/Datasets/paris-air-quality.csv'
airq_df = pd.read_csv(url)
airq_df
```

```python
for col in airq_df.columns:
    print(col)
airq_df.shape
#airq_df['date'].isnull().values.any()# remove each row with pclass = null
```

```python
final_airq = pd.DataFrame()
final_airq[['date','pm25','pm10','o3','no2','so2', 'co']] = pd.DataFrame([ x.split(';') for x in airq_df['date;pm25;pm10;o3;no2;so2; co'].tolist() ])
```

```python
final_airq
```

###Imputation with zero for airquality

```python
final_airq = final_airq.replace({' ': '0'}, regex=True)
final_airq
#final_airq.fillna(0)
#final_airq['co'].isnull().values.any()# remove each row with pclass = null
#final_airq = final_airq['co'].astype(float)
```

```python
for col in final_airq:
    if col!='date':
       final_airq[col] = final_airq[col].astype(float)
final_airq 
```

###We need to find the max of all column except datatime:This value would be AQI (label column)

```python
def calc_max(x):
  return x.max()
final_airq['AQI'] = final_airq[['pm25',	'pm10',	'o3',	'no2',	'so2',	'co']].apply(calc_max, axis =1)
final_airq
```
